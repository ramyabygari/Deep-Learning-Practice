```python
from bs4 import BeautifulSoup

tree = BeautifulSoup(r.text)
good_html = tree.prettify()
```

```python
soup=BeautifulSoup(good_html,'lxml')
soup1=BeautifulSoup(good_html,'html.parser')
soup2=BeautifulSoup(good_html,'html5lib')
```

```python
results=soup.find_all('span',attrs={"class":"short-desc"})
results1=soup1.find_all('span',attrs={"class":"short-desc"})
results2=soup2.find_all('span',attrs={"class":"short-desc"})
```

```python


##loop

records=[]
for result in results:
    date=result.find('strong').text[0:-1]+',2018'
    lie=result.contents[2]
    explanation=result.find('a').text
    url=result.find('a')['href']
    records.append((date,lie,explanation,url))
```

```python
records
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(records,columns=['data','lie','explanation','url'])
```

```python
df
```

```python
df['data']=df['data'].replace('\n','',regex=True)
```

```python
df['explanation']=df['explanation'].replace('\n','',regex=True)
df['explanation']=df['explanation']
```
